```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

rs = np.random.RandomState()

dim = [1] + list(range(100, 10001, 100))
rat = []

for d in dim:
    X = rs.randn(2000, d)
    norms = np.linalg.norm(X, axis=1)
    rat.append(norms.min() / norms.max())
```

```python
plt.figure()
plt.plot(dim, rat, marker='o')
plt.ylabel("min norm/max norm")
plt.xlabel("d")
plt.show()
```
